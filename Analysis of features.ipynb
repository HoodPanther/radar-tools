```python
from radar import *
from pointprocess.features import Features
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
city = 'CHARLOTTE'
computed = pd.HDFStore('{c}/computed.h5'.format(c=city))
p = computed['features_buffered_ratethresh10_minarea20_cont20']
computed.close()

TOP50 = pd.read_csv('{c}/TOP50_events.csv'.format(c=city), parse_dates=[0,1])
c = Radar(city, TOP50.t_begin[0], how='hdf5', make_rate=4, store='{c}/TOP50.h5')
ft = Features(p, c)
```

```python
ft.get_dirs4()
```

```python
fig = ft.plot_kde(pos=False, neg=False)
plt.savefig('{c}/Feature total KDE.png'.format(c=city))
```

```python
fig = ft.plot_kde(pos=True, neg=True, metrics=['area'], vmin=-.5, vmax=.5)
plt.savefig('{c}/Change in area KDE.png'.format(c=city))
```

```python
fig = ft.bearing_plot(N=32, bottom=10)
plt.savefig('{c}/Bearing for TOP50 events.png'.format(c=city))
```
